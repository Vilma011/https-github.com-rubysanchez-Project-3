# Google Maps Hospital Lat Long API

```python
#Rename dataframe and add Lat & Long placeholder columns
hospital_geocoords = Texas_Filtered_DF
hospital_geocoords["Latitude"] = ""
hospital_geocoords["Longitude"] = ""
hospital_geocoords.head()
```

```python
#Variables used to track responses
record = 1
set = 1

# create a params dict that will be updated with address information each iteration
from api_keys import gkey
params = {"key": gkey}

# Loop through the hospital_geocoords dataframe and run a lat/long search for each city
for index, row in hospital_geocoords.iterrows():
    url = "https://maps.googleapis.com/maps/api/geocode/json"
    
    Address = row['Address']
    City = row['City']
    State = row['State']
    Zip = row ['ZIP_Code']

    # update address key value
    params['address'] = f"{Address}+{City}+{State}+{Zip}"

    # make request
    hospitals_lat_lng = requests.get(url, params=params)


    
    # convert to json
    hospitals_lat_lng = hospitals_lat_lng.json()
    
#Search for/select first result from API and store in 'Lat & Long' columns
#If KeyError or IndexError present, print "Missing field/result... skipping." and continue iteration

    try:

#Prints retrieval information
        print(f'Processing Record {record} of Set {set} | {hospitals_lat_lng["results"][0]["geometry"]["location"]["lat"]} | {hospitals_lat_lng["results"][0]["geometry"]["location"]["lat"]}')
       
        hospital_geocoords.loc[index, "Latitude"] = hospitals_lat_lng["results"][0]["geometry"]["location"]["lat"]
        hospital_geocoords.loc[index, "Longitude"] = hospitals_lat_lng["results"][0]["geometry"]["location"]["lng"]
        
#Increases the record count by one each iteration
        record += 1
        
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")  
   
    #Increases set number by 1 when record number is greater than 50, record number is reset to 0
    if record > 100:
        set += 1
        record = 1

    time.sleep(1)

#Notification that all data is retrieved
print('''
-----------------------------
Data Retrieval Complete
-----------------------------''')  
```
